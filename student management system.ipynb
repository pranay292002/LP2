```python
import csv

def add_student():
    name = input("Enter name: ")
    age = input("Enter age: ")
    gender = input("Enter gender: ")
    email = input("Enter email: ")
    phone = input("Enter phone number: ")
    with open('students.csv', 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([name, age, gender, email, phone])
    print("Student added successfully.")

def view_students():
    with open('students.csv', 'r') as file:
        reader = csv.reader(file)
        for row in reader:
            print(row)

def search_student():
    name = input("Enter name of student to search: ")
    with open('students.csv', 'r') as file:
        reader = csv.reader(file)
        for row in reader:
            if name == row[0]:
                print(row)
                break
        else:
            print("Student not found.")

def update_student():
    name = input("Enter name of student to update: ")
    field = input("Enter field to update (name/age/gender/email/phone): ")
    new_value = input("Enter new value: ")
    with open('students.csv', 'r') as file:
        reader = csv.reader(file)
        rows = list(reader)
    with open('students.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        for row in rows:
            if name == row[0]:
                if field == 'name':
                    row[0] = new_value
                elif field == 'age':
                    row[1] = new_value
                elif field == 'gender':
                    row[2] = new_value
                elif field == 'email':
                    row[3] = new_value
                elif field == 'phone':
                    row[4] = new_value
            writer.writerow(row)
    print("Student updated successfully.")

def delete_student():
    name = input("Enter name of student to delete: ")
    with open('students.csv', 'r') as file:
        reader = csv.reader(file)
        rows = list(reader)
    with open('students.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        for row in rows:
            if name != row[0]:
                writer.writerow(row)
    print("Student deleted successfully.")

while True:
    print("\nStudent Management System")
    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Update Student")
    print("5. Delete Student")
    print("6. Quit")
    choice = input("Enter your choice: ")
    if choice == '1':
        add_student()
    elif choice == '2':
        view_students()
    elif choice == '3':
        search_student()
    elif choice == '4':
        update_student()
    elif choice == '5':
        delete_student()
    elif choice == '6':
        break
    else:
        print("Invalid choice. Please try again.")

```
